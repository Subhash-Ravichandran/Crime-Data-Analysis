```python
import pandas as pd
import numpy as np
```

```python
# reading xlsx file
df_sample=pd.read_excel('/content/Sample Crime Dataset.xlsx')
```

```python
df_sample.head()
```

```python
df_sample.shape
```

```python
df_sample.info()
```

```python
df_sample.isnull().sum()
```

```python
#filling null values for categorical data using mode
df_sample["Location Description"].fillna(df_sample["Location Description"].mode()[0],inplace=True)
df_sample["Ward"].fillna(df_sample["Ward"].mode()[0],inplace=True)
df_sample["Community Area"].fillna(df_sample["Community Area"].mode()[0],inplace=True)
```

```python
#replacing null values for locations with "unknown"
df_sample["X Coordinate"].replace(np.nan,"unknown",inplace=True)
df_sample["Y Coordinate"].replace(np.nan,"unknown",inplace=True)
df_sample["Location"].replace(np.nan,"unknown",inplace=True)
df_sample["Latitude"].replace(np.nan,"unknown",inplace=True)
df_sample["Longitude"].replace(np.nan,"unknown",inplace=True)
```

```python
df_sample.isnull().sum()
```

```python
df_sample["Primary Type"].unique()
```

```python
# seperating date and time from same column
df_sample['Date'] = pd.to_datetime(df_sample['Date'])

df_sample['Crime_Date'] = df_sample['Date'].dt.date
df_sample['Crime_Time'] = df_sample['Date'].dt.time
```

```python
# seperating updated date and time from same column
df_sample['Updated On'] = pd.to_datetime(df_sample['Updated On'])

df_sample['Updated_Date'] = df_sample['Updated On'].dt.date
df_sample['Updated_Time'] = df_sample['Updated On'].dt.time
```

```python
# droping unnecessary columns
df_sample.drop(['Date'],axis=1,inplace=True)
df_sample.drop(['Updated On'],axis=1,inplace=True)
```

```python
df_sample.head()
```

```python
df_sample.info()
```

```python
# converting the final dataframe to csv file
df_sample.to_csv('SampleData_final.csv',index=False)
```
